```python
import pandas as pd
from pytrends.request import TrendReq

pytrend=TrendReq()
```

```python
pytrend.build_payload(kw_list=['vaccine'])

df=pytrend.interest_by_region()
df.head(100)
```

```python
df.columns
```

```python
df.reset_index().plot(x='geoName', y='vaccine',figsize=(120,50), kind='bar')
```

```python
df=pytrend.trending_searches(pn='india')
df.head(20)
```

```python
df=pytrend.today_searches(pn='IN')
df.head(20)
```

```python
df=pytrend.top_charts(2020,hl='en-IN',tz=300, geo='GLOBAL')
df.head(10)
```
